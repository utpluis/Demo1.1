```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from google.colab import files
```

```python
df = pd.read_excel('ventas_100_filas.xlsx')

# Convertir la columna de fecha
df['Fecha de Venta'] = pd.to_datetime(df['Fecha de Venta'])

# Agrupar por fecha y sumar cantidad vendida
ventas_diarias = df.groupby('Fecha de Venta')['Cantidad'].sum().reset_index()

# Renombrar columnas para Prophet
ventas_diarias.columns = ['ds', 'y']  # Prophet requiere 'ds' para fecha y 'y' para valor

# PASO 5: VISUALIZAR DATOS HISTÓRICOS
plt.figure(figsize=(12, 5))
plt.plot(ventas_diarias['ds'], ventas_diarias['y'], marker='o')
plt.title('Ventas diarias (histórico)')
plt.xlabel('Fecha')
plt.ylabel('Cantidad Vendida')
plt.grid(True)
plt.show()

# PASO 6: CREAR Y ENTRENAR EL MODELO PROPHET
modelo = Prophet(daily_seasonality=True)
modelo.fit(ventas_diarias)

# PASO 7: CREAR FECHAS FUTURAS PARA PRONÓSTICO
futuro = modelo.make_future_dataframe(periods=15)  # 15 días hacia adelante

# PASO 8: GENERAR PRONÓSTICO
forecast = modelo.predict(futuro)

# PASO 9: VISUALIZAR PROYECCIÓN
fig = modelo.plot(forecast)
plt.title("Proyección de ventas para los próximos 15 días")
plt.xlabel("Fecha")
plt.ylabel("Cantidad proyectada")
plt.grid(True)
plt.show()

# PASO 10: EXPORTAR PRONÓSTICO
# Filtrar solo los próximos 15 días de pronóstico
proyeccion_futura = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(15)
proyeccion_futura.rename(columns={
    'ds': 'Fecha',
    'yhat': 'Pronóstico',
    'yhat_lower': 'Intervalo Inferior',
    'yhat_upper': 'Intervalo Superior'
}, inplace=True)

# Guardar en Excel
proyeccion_futura.to_excel('proyeccion_ventas.xlsx', index=False)

# Descargar archivo
files.download('proyeccion_ventas.xlsx')
```
