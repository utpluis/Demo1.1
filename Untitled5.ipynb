```python
# Paso 2: Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from google.colab import files
```

```python
# Paso 3: Subir archivo Excel (ventas_10000_filas.xlsx)
uploaded = files.upload()
```

```python
# Paso 4: Cargar y preparar la serie de tiempo
df = pd.read_excel('ventas_10000_filas.xlsx')
df['Fecha de Venta'] = pd.to_datetime(df['Fecha de Venta'])
```

```python
# Agrupar por fecha y sumar ventas
ventas_diarias = df.groupby('Fecha de Venta')['Cantidad'].sum().reset_index()
ventas_diarias.columns = ['ds', 'y']  # Renombrar para Prophet
```

```python
# Paso 5: Visualizar datos históricos
plt.figure(figsize=(14, 5))
plt.plot(ventas_diarias['ds'], ventas_diarias['y'], marker='o', linestyle='-', color='darkblue')
plt.title('Histórico de Ventas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Cantidad Vendida')
plt.grid(True)
plt.show()
```

```python
# Paso 6: Crear y entrenar modelo Prophet
modelo = Prophet(daily_seasonality=True)
modelo.fit(ventas_diarias)
```

```python
# Función para generar y exportar pronósticos escalables
def generar_proyeccion(periodos, nombre_archivo):
    futuro = modelo.make_future_dataframe(periods=periodos)
    forecast = modelo.predict(futuro)
```

```python
def generar_proyeccion(periodos, nombre_archivo):
    futuro = modelo.make_future_dataframe(periods=periodos)
    forecast = modelo.predict(futuro)

    # Visualizar
    fig = modelo.plot(forecast)
    ax = fig.gca()
    ax.set_title(f"Proyección de ventas ({periodos} días)")
    ax.set_xlabel("Fecha")
    ax.set_ylabel("Cantidad Proyectada")
    ax.grid(True)
    plt.tight_layout()
    plt.show()

    # Exportar últimos n días
    proyeccion = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(periodos).copy()
    proyeccion = proyeccion.rename(columns={
        'ds': 'Fecha',
        'yhat': 'Pronóstico',
        'yhat_lower': 'Inferior',
        'yhat_upper': 'Superior'
    })
    proyeccion.to_excel(nombre_archivo, index=False)
    files.download(nombre_archivo)
```

```python
def exportar_proyeccion(periodos, nombre_archivo):
    futuro = modelo.make_future_dataframe(periods=periodos)
    forecast = modelo.predict(futuro)

    proyeccion = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(periodos).copy()

    proyeccion = proyeccion.rename(columns={
        'ds': 'Fecha',
        'yhat': 'Pronóstico',
        'yhat_lower': 'Inferior',
        'yhat_upper': 'Superior'
    })

    proyeccion.to_excel(nombre_archivo, index=False)
    from google.colab import files
    files.download(nombre_archivo)
```
